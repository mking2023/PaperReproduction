```python
import pandas as pd
import numpy as np
```

```python
file_path = '../Tools.csv'
```

```python
df = pd.read_csv(file_path, sep='\t')
df
```

```python
uid = []
uid_train = []
uid_valid = []
uid_test = []

for group in df.groupby(by='uid'):
    cur_uid = group[0]
    cur_item_lst = group[1]['iid'].tolist()
    cur_valid = cur_item_lst[-2]
    cur_test = cur_item_lst[-1]
    cur_train = cur_item_lst[:-2]
    uid_train.append(cur_train)
    uid_valid.append(cur_valid)
    uid_test.append(cur_test)
    uid.append(cur_uid)
```

```python
train_df = pd.DataFrame({
    'uid': uid,
    'iid' : uid_train
})

valid_df = pd.DataFrame({
    'uid': uid,
    'iid': uid_valid
})

test_df = pd.DataFrame({
    'uid': uid,
    'iid' : uid_test
})
```

```python
train_df = train_df.explode('iid').reset_index(drop=True)
train_df
```

```python
train_df['label'] = 1
valid_df['label'] = 1
test_df['label'] = 1
```

```python
dataset_name = 'Tools'
train_df.to_csv(f'./{dataset_name}.train.csv', index=False, sep='\t')
valid_df.to_csv(f'./{dataset_name}.validation.csv', index=False, sep='\t')
test_df.to_csv(f'./{dataset_name}.test.csv', index=False, sep='\t')
```
